**Impute missing monthly values in a time series dataset**

```python
##  Use this code when the time series has missing values and you need to
## impute the missing months (e.g., source: OJ014).

import pandas as pd

# Load data
data = pd.read_csv('OJ014.csv', usecols=['Date (UTC)', 'Flux'])

# Rename columns
data = data.rename(columns={'Date (UTC)': 'date', 'Flux': 'y'})

data = data.iloc[:168].copy() # Number of rows 168

# Convert to datetime and set monthly period index
data['date'] = pd.to_datetime(data['date'], format='%m/%d/%Y')
data['date'] = data['date'].dt.to_period('M')
data = data.set_index('date')

# Reindex to complete monthly range
full_date_range = pd.period_range(start=data.index.min(), end=data.index.max(), freq='M')
data = data.reindex(full_date_range)

# Check missing values before imputation
print("Missing values before imputation:")
print(data.isnull().sum())

# Impute missing values using linear interpolation
data_imputed = data.interpolate(method='linear')

# Check missing values after imputation
print("\nMissing values after imputation:")
print(data_imputed.isnull().sum())

# Save full imputed dataset to CSV
data_imputed.to_csv('OJ014_imputed.csv')
print("\nFull imputed dataset saved to 'OJ014_imputed.csv'")

# Extract and save only the imputed values for originally missing months
missing_months = data[data['y'].isnull()]
imputed_values = data_imputed.loc[missing_months.index]

# Save imputed values for missing months to text file
imputed_values.to_csv('imputedMissingValues.txt', sep='\t', header=['ImputedValue'])
print("Imputed missing values saved to 'imputedMissingValues.txt'")

# Display imputed values
print("\nImputed Values for Missing Months:")
print(imputed_values)
```
